```python
## String Functions

# Initial string
s = "Skill Morph"

# 1. upper()
print(s.upper())  # Converts all characters to uppercase
# 2. lower()
print(s.lower())  # Converts all characters to lowercase
# 3. partition()
print(s.partition(' '))  # Splits the string at the first occurrence of ' '
# 4. replace()
print(s.replace('Morph', 'Master'))  # Replaces 'Morph' with 'Master'
# 5. find()
print(s.find('Morph'))  # Returns the index of 'Morph' in the string, or -1 if not found
# 6. rstrip()
print(s.rstrip())  # Removes trailing spaces from the string
# 7. split()
print(s.split())  # Splits the string into a list at spaces
# 8. startswith()
print(s.startswith('Skill'))  # Checks if the string starts with 'Skill'
# 9. isnumeric()
print(s.isnumeric())  # Checks if all characters in the string are numeric (returns False)
# 10. index()
print(s.index('Morph'))  # Returns the index of the first occurrence of 'Morph'
```

```python
#H9 Write a program to find the key in a dictionary that has the
maximum value.
```

```python
data = {'a': 10, 'b': 20, 'c': 15}

max_key = max(data, key=data.get)

print(max_key)


data = {
    "A": 10,
    "B": 50,
    "C": 30,
    "D": 80
}

max_key = max(data, key=data.get)

print("Key with maximum value:", max_key)
print("Maximum value:", data[max_key])

```
