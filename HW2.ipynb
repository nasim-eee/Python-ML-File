## H9: Write a program that asks the user to enter three numbers and determines which one is the largest.

```python
while True:
  print("type a for entering numbers, q for quit")
  user_input=str(input())
  if user_input.lower()=="q":
    print("quit the program")
    break
  elif user_input.lower()=="a":
    print("enter the numbers")
    num1=float(input("enter first number: "))
    num2=float(input("enter second number: "))
    num3=float(input("enter third number: "))
    if num1>num2 and num2>num3:
        print(f"the largest number is num1: {num1}")
    elif num2>num1 and num2>num3:
        print(f"the largest number is num2: {num2}")
    else:
        print(f"the largest number is num3: {num3}")
    break
  else:
    print("invalid input! please type again")


print("done!")
```

# H11: Write a program that prompts the user to input a number N and calculates the sum of all numbers from 1 to N that are divisible by 3 or 5.

```python
sum=0
N=int(input("Enter a number N: "))
for x in range(1, N+1):
  if x%3==0 or x%5==0:
    print(x)
    sum=sum+x
    print(f" sum is {sum}")
print(f"final sum of all that numbers is {sum}")
```
